## 1.4.1 파이썬 스크립트 파일

## 1.4.2 클래스  
변수와 함수를 저장하고 사용하는 틀, 멤버와 메서드를 정의한다.

```python
class Man:
  def __init__(self,name):#클래스의 인스턴스가 만들어질때 한번만 호출됨
    self.name=name
    print("Initialized")
  def hello(self):
    print("Hello "+self.name+"!")
  def goodbye(self):
    print("Goodbye "+self.name+"!")
m=Man("David")
m.hello()
m.goodbye()
```

## 1.6 matplotlib  
그래프를 그려주는 라이브러리

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,6,0.1)#0~6까지 0.1단위로 생성
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

## 1.6.2 pyplot의 기능  
제목, x,y축등의 레이블 이름 표시등과 같은 기능도 제공한다.

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,6,0.1)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label='sin')
plt.plot(x,y2,ls="--",label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin and cos")
plt.legend()#범례 보이기
plt.show()
```
