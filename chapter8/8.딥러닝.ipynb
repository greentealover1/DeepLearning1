# 8.1 더 깊게  
딥러닝:층을 깊게 한 심층신경망  
현재 이 코드의 목표:실제로 MNIST데이터셋의 손글씨 숫자인식에 도전하기  

## 8.1.1 더 깊은 신경망으로  
손글씨 숫자 인식하는 심층 CNN의 처리과정: 손글씨->conv->ReLU->conv->ReLU->pool계층을 거치고 Affine0>Relu->dropout을 거친 후에 Softmax층을 통과  

신경망의 특징  
- 3x3크기의 작은 필터를 사용한 합성곱 계층  
- 활성화함수:ReLU  
- 완전연결계층 뒤에 드롭아웃 계층 사용  
- Adam을 사용해 최적화 및 가중치 매개변수 갱신   
- 가중치 초깃값:'He의 초깃값'사용  

## 8.1.2 정확도를 더 높이려면?  
결국 뉴럴신경망, 딥러닝,CNN기법등을 사용하고 '앙상블 학습','학습률 감소','데이터 확장' 등이 정확도향상에 좋다  
데이터 확장-입력 이미지를 알고리즘을 동원해 '인위적'으로 확장한다(by 입력 이미지를 회전 또는 세로로 이동하는 등 미세한 변화주기)   


## 8.1.3 깊게 하는 이유
층을 깊게 하는 이유(중요성):
1. 대규모 이미지 인식 대회의 결과에서 상위권기법들은 대부분 딥러닝 기반  
2. 신경망의 매개변수 수가 줄어든다(적은 매개변수로 같은 또는 그 이상의 표현력을 달성할 수 있다)  
이때, 주의할점은 출력데이터의 각 노드가 입력데이터의 어느 영역으로부터 계산되는가 하는 것임  
ex.5x5합성곱연산1회는 3x3의 합성곱연산을 2회 수행하여 대체가능.(전자의 매개변수는 25개, 후자는 18개이다)  

3. 학습의 효율성 극대화(층을 깊게 하면, 학습데이터양도 줄여 학습을 고속으로 수행 가능)  
4. 학습해야할 문제를 계층적으로 분해할 수 있음  
4. 정보를 계층적으로 전달가능 (ex.에지를 추출한 층의 다음층은 에지정보를 쓸 수 있고, 고도의 패턴을 효과적으로 학습할 수 있게됨)

# 8.2 딥러닝의 초기역사  

## 8.2.1 이미지넷  
이미지넷:100만장이 넘는 이미지를 담고 있는 데이터셋  
다양한 종류의 이미지와 각 이미지에는 레이블(클래스 이름)이 붙음  

## 8.2.2 VGG  
VGG:합성곱계층과 풀링계층으로 구성되는 '기본적'인 CNN이다  
비중있는 층(합성곱계층, 완전연결계층)은 모두 16층(19층)으로 심화한게 특징  
- 합성곱계층은 2~4회 연속으로 풀링계층을 두어 크기를 절반으로 줄이는 처리 반복.마지막에는 완전연결계층을 통과시켜 결과를 출력한다  

## 8.2.3 GoogLeNet  
특징:  
1. GoogLeNet은 세로 방향 깊이뿐 아니라 가로방향도 깊다 (GoogLeNet은 가로방향에 '폭'이 존재-인셉션 구조)  
cf)인셉션구조:크기가 다른 필터를 여러개 적용하여 결과를 결합한다  
- 인셉션구조를 하나의 구성요소(빌딩블록)으로 사용하는 것이 'GoogLeNet'  
2. 1x1크기의 필터를 사용한 합성곱계층을 많은 곳에서 사용.1x1합성곱 연산은 채널쪽으로 크기를 줄이는 것으로 매개변수제거와 고속처리에 기여  

## 8.2.4 ResNet  
딥러닝에서는 층이 지나치게 깊어지면 학습이 잘 안되고 성능이 떨어지는 경우도 은근히 있었음  
ResNet의 특이점: 스킵연결을 도입  

스킵연결: 입력데이터를 합성곱 계층을 건너뚸서 출력에 바로 더하는 구조!!!  
- 스킵연결은 층이 깊어져도 학습을 효율적으로 수행 가능: 역전파때 스킵연결이 신호감쇠를 막기 때문  
-> ResNet은 합성곱계층을 2개층마다 건너뚸서 층을 깊게 함  

전이 학습transfer learning : 학습된 가중치를 다른 신경망에 복사한 후,재학습을 수행하는 것. 보유한 데이터셋이 적을 때 유용  

# 8.3 더 빠르게(딥러닝 고속화)  
cpu만으로 딥러닝을 처리하기에는 부족->따라서 딥러닝 프레임워크 대부분은 GPU를 활용해 대량의 연산을 고속으로 처리가능  

## 8.3.1 풀어야 할 숙제  
AlexNet에서는 오랜 시간을 합성곱계층에서 소요!- 결국 학습시에서도 마찬가지로 합셩곱계층에서 많은 시간 소비  

## 8.3.2 GPU를 활용한 고속화  
GPU컴퓨팅의 목적:GPU의 고속의 병렬수치연산을 다양한 용도로 활용하는 것  
CF)GPU:대량 병렬연산을 고속으로 수행 vs CPU:연속적인 복잡한 계산을 잘 처리  
엔비디아의 GPU컴퓨팅용 통합개발환경 CUDA가 존재(가장 많이 활용됨)  
ex)cuDNN은 CUDA위에서 동작하는 라이브러리로 딥러닝에 최적화된 함수등이 구현되어있다  

## 8.3.3 분산 학습  
GPU로 딥러닝 연산을 가속할 수 있음.But, 많은 시행착오를 동반함->딥러닝 학습을 수평확장하자는 아이디어가 중요해짐  
1회 학습에 걸리는 시간을 단축하기 위해 학습을 수평확장하기 위해 다수의 GPU와 컴퓨터에 분산한다.  
텐서플로의 경우 분산학습의 효과는 GPU100개 사용시 1개일때보다 56배 더 빨라짐.  

## 8.3.4 연산정밀도와 비트줄이기  
계산 능력외에도 메모리 용량, 버스 대역폭등이 딥러닝 고속화에 병목이 될 수 있다  
메모리 용량면에서는 대량의 가중치 매개변수와 중간데이터를 메모리에 저장해야함  
->네트워크로 주고받는 데이터의 bit수를 최소화한다  
cf)많은 bit를 사용할 수록 오차는 줄지만 계산비용,메모리사용량이 늘고 버스대역폭에 부담이 높아짐  
But,딥러닝에서는 일반적으로 16bit반정밀도를 표준적으로 이용  

# 8.4 딥러닝의 활용  
딥러닝은 사물 인식뿐만 아니라 온갖문제에 적용할 수 있다  
이미지, 음성, 자연어 등 수많은 분야에서 딥러닝은 뛰어난 성능을 발휘

## 8.4.1 사물검출  
사물검출:이미지속에 담긴 사물의 위치, 종류(클래스)를 알아내는 기술  
- 사물검출은 사물인식이 이미지 전체를 대상으로 한 것에 비해, 이미지 어딘가에 있을 사물위치도 알아야 해서 더 어려움  

r-cnn의 처리흐름  
1. 입력이미지
2. 후보영역 추출  
3. CNN특징계산  
4. 영역 분류  
R-CNN은 사물이 위치한 영역을 찾아내고, 추출한 각 영역에 CNN을 적용해서 클래스를 분류(후보영역추출,CNN특징계산으로 구성됨)  

cf)후보영역추출까지도 CNN으로 처리하는 Faster R-CNN기법도 등장  

## 8.4.2 분할  
분할:이미지를 픽셀 수준에서 분류하는 문제  
픽셀단위로 객체마다 채색된 지도데이터를 사용해 학습->추론할때 입력이미지의 모든 픽셀을 분류  

신경망을 이용해 분할하는 가장 단순한 방법은 모든 픽셀 각각을 추론하는 것  
ex)어떤 직사각형 영역의 중심 픽셀의 클래스를 분류하는 신경망을 만들어서 모든 픽셀을 대상으로 하나씩 추론작업을 실행.FCN(단한번의 forward처리로 모든 픽셀 클래스를 분류)를 통해 긴 시간이 소요된다는 문제를 해결해준다  

일반적인 CNN이 완전연결 계층을 이용하는 반면 FCN에서는 이를 '같은 기능을 하는 합성곱 계층'으로  
바꾼다. 또, 마지막 출력까지 공간 볼륨을 유지한 채 처리가능   

CF)입력 크기가 32*10*10인 데이터의 완전 연결 계층은 필터 크기가 32*10*10인 합성곱 계층으로 대체할 수 있다.  
CF)완전연결 계층의 출력 노드가 100개라면 이러한 필터 100개를 사용하면 된다.  
즉, 완전연결계층은 같은 일을 수행하는 합성곱계층으로 대체가능!

## 8.4.3 사진 캡션 생성  
NIC모델:심층 CNN과 자연어를 다루는 순환신경망 RNN으로 구성된다.  (RNN은 순환적 관계를 갖는 신경망으로 자연어나 시계열 데이터등의 연속된 데이터를 다룰때 많이 활용된다)   
Nic는 CNN으로 사진에서 특징 추출->특징을 RNN에 넘김->RNN은 CNN이 추출한 특징을 초깃값으로 해서 텍스트를 순환적으로 생성  

Multimodal processing멀티모달처리:사진, 자연어와 같은 여러종류의 정보를 조합하고 처리하는 것  

# 8.5 딥러닝의 미래  
## 8.5.1 이미지 스타일 (화풍)변환  
ex.네트워크의 중간데이터가 콘텐츠 이미지의 중간데이터와 비슷하도록 학습하여 입력이미지를 콘텐츠이미의 형태를 흉내낼 수 있음  
ex.스타일 이미지의 화풍을 흡수하기 위해 '스타일 행렬'개념 도입  

## 8.5.2 이미지 생성  
이미지 스타일 변환에는 새로운 그림을 생성하려면 이미지 두장을 입력해야했음...  
대량의 이미지를 사용하여 학습하긴 하지만, 학습이 끝난 후에는 아무런 입력 이미지 없이도 새로운 그림을 그려낼 수 있도록 연구진행중(ex.DCGAN)  

DCGAN:이미지를 생성하는 과정을 모델화한다  
모델을 대량의 이미지를 사용해 학습하고, 학습이 끝나면 그 모델을 이용해서 새로운 그림을 생성할 수 있음  
- 특징:생성자(진짜와 똑같은 이미지 생성하고 이후에 더 정교한 가짜 이미지 생성기술학습)와 식별자(생성자가 생성한 이미지인지, 아니면 실제로 촬영된 이미지인지)를 판정한다  
즉, 생성자와 식별자라는 2개의 신경망을 이용하는 것!  

## 8.5.3 자율 주행  
자율주행에서 가장 주의해야할 요소: 주위환경을 올바르게 인식하는 기술이 가장 중요(시시각각 변하는 환경과 다양한 사물을 올바르게 인식하기 어려워서)  
- 자율주행분야에서 딥러닝의 활약:SegNet은 주변환경을 정확하게 인식해낸다  

## 8.5.4 Deep Q-Network(강화학습)  
강화학습:에이전트가 환경에 맞게 행동을 선택하고, 그 행동에 의해 환경이 변한다. 환경이 변하면 에이전트가 보상을 받고 더 나은 보상을 받는쪽으로 에이전트가 향후 행동을 수행하게 된다  

딥러닝을 이용한 강화학습(Deep Q-Network):  
Q학습은 강화학습 알고리즘을 기초로 한다  
Q학습에서는 최적행동가치 함수로 최적인 행동을 정함->딥러닝으로 비슷하게 흉내낸것이 DQN
